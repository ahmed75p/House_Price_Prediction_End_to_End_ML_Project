### **Outliers**

```python
def detect_outliers(data , column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    Lower_bound = Q1 - 1.5 *IQR
    Upper_boun = Q3 + 1.5 *IQR

    outliers = data[(data[column]<Lower_bound) | (data[column]>Upper_boun)]
    return outliers

# Detect outliers in price
out =detect_outliers(house_data_cleaning, 'price')
print(f"Number of outliers in price: {out.shape[0]}")

out =detect_outliers(house_data_cleaning, 'bedrooms')
print(f"Number of outliers in bedrooms : {out.shape[0]}")

out =detect_outliers(house_data_cleaning, 'sqft_living')
print(f"Number of outliers in sqft_living: {out.shape[0]}")

out =detect_outliers(house_data_cleaning, 'sqft_lot')
print(f"Number of outliers in sqft_lot: {out.shape[0]}")

out =detect_outliers(house_data_cleaning, 'sqft_above')
print(f"Number of outliers in sqft_above: {out.shape[0]}")

out =detect_outliers(house_data_cleaning, 'sqft_basement')
print(f"Number of outliers in sqft_basement: {out.shape[0]}")
```

```python
Q1_price = house_data_cleaning['price'].quantile(0.25)
Q3_price = house_data_cleaning['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_bound_price = Q1_price - 1.5 * IQR_price
upper_bound_price = Q3_price + 1.5 * IQR_price
house_data_cleaning = house_data_cleaning[
    (house_data_cleaning['price'] >= lower_bound_price) & (house_data_cleaning['price'] <= upper_bound_price)]


       
Q1_sqft_living = house_data_cleaning['sqft_living'].quantile(0.25)
Q3_sqft_living = house_data_cleaning['sqft_living'].quantile(0.75)
IQR_sqft_living = Q3_sqft_living - Q1_sqft_living
lower_bound_sqft_living = Q1_sqft_living - 1.5 * IQR_sqft_living
upper_bound_sqft_living = Q3_sqft_living+ 1.5 * IQR_sqft_living
house_data_cleaning = house_data_cleaning[
    (house_data_cleaning['sqft_living'] >= lower_bound_sqft_living) & (house_data_cleaning['sqft_living'] <= upper_bound_sqft_living)]

Q1_sqft_lot = house_data_cleaning['sqft_lot'].quantile(0.25)
Q3_sqft_lot = house_data_cleaning['sqft_lot'].quantile(0.75)
IQR_sqft_lot = Q3_sqft_lot - Q1_sqft_lot
lower_bound_sqft_lot = Q1_sqft_lot- 1.5 * IQR_sqft_lot
upper_bound_sqft_lot = Q3_sqft_lot + 1.5 * IQR_sqft_lot
house_data_cleaning = house_data_cleaning[
    (house_data_cleaning['sqft_lot'] >= lower_bound_sqft_lot) & (house_data_cleaning['sqft_lot'] <= upper_bound_sqft_lot)]

Q1_sqft_above = house_data_cleaning['sqft_above'].quantile(0.25)
Q3_sqft_above = house_data_cleaning['sqft_above'].quantile(0.75)
IQR_sqft_above = Q3_sqft_above- Q1_sqft_above
lower_bound_sqft_above = Q1_sqft_above - 1.5 * IQR_sqft_above
upper_bound_sqft_above = Q3_sqft_above+ 1.5 * IQR_sqft_above
house_data_cleaning = house_data_cleaning[
    (house_data_cleaning['sqft_above'] >= lower_bound_sqft_above) & (house_data_cleaning['sqft_above'] <= upper_bound_sqft_above)]

Q1_sqft_basement = house_data_cleaning['sqft_basement'].quantile(0.25)
Q3_sqft_basement= house_data_cleaning['sqft_basement'].quantile(0.75)
IQR_sqft_basement = Q3_sqft_basement - Q1_sqft_basement
lower_bound_sqft_basement = Q1_sqft_basement - 1.5 * IQR_sqft_basement
upper_bound_sqft_basement = Q3_sqft_basement + 1.5 * IQR_sqft_basement
house_data_cleaning = house_data_cleaning[
    (house_data_cleaning['sqft_basement'] >= lower_bound_sqft_basement) & (house_data_cleaning['sqft_basement'] <= upper_bound_sqft_basement)]

Q1_bedrooms = house_data_cleaning['bedrooms'].quantile(0.25)
Q3_bedrooms= house_data_cleaning['bedrooms'].quantile(0.75)
IQR_bedrooms = Q3_bedrooms - Q1_bedrooms
lower_bound_bedrooms = Q1_bedrooms - 1.5 * IQR_bedrooms
upper_bound_bedrooms = Q3_bedrooms + 1.5 * IQR_bedrooms
house_data_cleaning = house_data_cleaning[
    (house_data_cleaning['bedrooms'] >= lower_bound_bedrooms) & (house_data_cleaning['bedrooms'] <= upper_bound_bedrooms)]
```
